```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
print('modules are imported.')
```

```python
cd = pd.read_csv('time_series_covid19_confirmed_global.csv')
cd.head(10)
```

```python
cd.shape
```

```python
df = cd.drop(["Lat","Long"],axis=1)
df.head(10)
```

```python
cd_aggregated = cd.groupby("Country/Region").sum()
cda = cd_aggregated.drop(["Lat","Long"],axis=1)
cda.head(10)
```

```python
cda.shape
```

```python
cda.loc["China"].plot()
cda.loc["Italy"].plot()
cda.loc["India"].plot()
plt.legend()
```

```python
cda.loc["China"][:3].plot()
```

```python
cda.loc["China"].diff().plot()
```

```python
cda.loc["China"].diff().max()
```

```python
cda.loc["India"].diff().max()
```

```python
countries = list(cda.index)
max_inf = []
for c in countries :
    max_inf.append(cda.loc[c].diff().max())
cda["max_inf_rate"] = max_inf
cda.head(10)
```

```python
corona_data = pd.DataFrame(cda["max_inf_rate"])
corona_data.head()
```
