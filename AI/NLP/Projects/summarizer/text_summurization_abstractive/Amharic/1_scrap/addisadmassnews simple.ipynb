```
#https://stackoverflow.com/questions/50270232/scrape-all-of-sublinks-of-a-website-recursively-in-python-using-beautiful-soup
from scrapy.spiders import CrawlSpider
from scrapy.spiders import Rule
from scrapy.linkextractors import LinkExtractor

from scrapy import Item
from scrapy import Field


class UrlItem(Item):
    url = Field()


class WikiSpider(CrawlSpider):
    #name = 'wiki'
    #allowed_domains = ['dw.com/am']
    start_urls = ['https://www.ethiopianregistrar.com/amharic/']

    rules = (
        Rule(LinkExtractor(), callback='parse_url'),
    )

    def parse_url(self, response):
        item = UrlItem()
        item['url'] = response.url

        return item
```

```
#https://stackoverflow.com/questions/40856730/how-to-run-scrapy-project-in-jupyter
import scrapy
from scrapy.crawler import CrawlerProcess

process = CrawlerProcess({
    'USER_AGENT': 'Mozilla/4.0 (compatible; MSIE 7.0; Windows NT 5.1)',
    'FEED_FORMAT': 'CSV', 
    'FEED_URI': 'file:///content/am.csv',
})

process.crawl(WikiSpider)
process.start() # the script will block here until the crawling is finished
```

```
from bs4 import BeautifulSoup
import requests
import pandas as pd

main_link = "https://www.addisadmassnews.com/index.php?option=com_k2&view=item&id="

output = [] 

start = 23000 #12001  #700
end = 24863
save_period = 100 #1000

for num in range(start,end):
  
  try:
    local = main_link + str(num)
    response = requests.get(local)
    data = response.text
    soup = BeautifulSoup(data, 'lxml')

    title = soup.find("h2", {"class": "itemTitle"})
    content = soup.find("div", {"class": "itemFullText"})
    
    row = {}
    row["title"] = title.text
    row["content"] = content.text
    row["link"] = str(local)
    row["id"] = str(num)

    output.append(row)
    print(num)
    if num % save_period == 0 and len(output) > 0 and num != start:
      print("saving")
      pd.DataFrame(output).to_csv("/content/drive/My Drive/Amhari/news_amhari_addisadmassnews.csv", mode='a',index = False,encoding='utf-8')
      output = []
  except:
    pass


#df = pd.DataFrame(output)
#df.to_csv("/content/drive/My Drive/Amhari/news_amhari.csv", encoding='utf-8')
```

```
from bs4 import BeautifulSoup
import requests

gloab_links = set()
def get_links(url):
    response = requests.get(url)
    data = response.text
    soup = BeautifulSoup(data, 'lxml')

    links = set()
    for link in soup.find_all('a'):
        link_url = link.get('href')
        print(link_url)
        if link_url is not None and link_url.startswith('https://www.zehabesha.com/amharic/'):
            links.add(link_url + '\n')

    write_to_file(links)
    return links


def write_to_file(links):
    with open('zehabesha_1.txt', 'a') as f:
        f.writelines(links)


def get_all_links(url):
    for link in get_links(url):
      try:
        if link not in gloab_links\
         and link != 'http://www.litespeedtech.com/error-page'\
         and  link.startswith('https://www.zehabesha.com/amharic/'):
          gloab_links.add(link)
          get_all_links(link)
      except:
        continue


r = 'https://www.zehabesha.com/amharic/'
write_to_file([r])
get_all_links(r)
```
