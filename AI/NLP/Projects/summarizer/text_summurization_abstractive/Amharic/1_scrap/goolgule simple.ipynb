```
from bs4 import BeautifulSoup
import requests

def get_more_links(link):
  response = requests.get(link)
  data = response.text
  soup = BeautifulSoup(data, 'lxml')
  more_links = soup.find_all("a", {"class": "entry-image-link"})
  more_links = [m["href"] for m in more_links]
  return more_links

def get_title_content(link):
  response = requests.get(link)
  data = response.text
  soup = BeautifulSoup(data, 'lxml')
  [s.extract() for s in soup('style')]
  [s.extract() for s in soup('script')]
  title = soup.find("h1",  {"class": "entry-title"})
  content = soup.find("div",  {"class": "pf-content"})
  title = title.text
  content = content.text
  content = " ".join(content.split(" ")[:500])
  return title,content

def see_next_page(link):
  response = requests.get(link)
  data = response.text
  soup = BeautifulSoup(data, 'lxml')
  [s.extract() for s in soup('style')]
  [s.extract() for s in soup('script')]
  next_page = soup.find("div",  class_="pagination-next alignright")
  if next_page!= None:
    next_page = next_page.find("a")["href"]
    return True
  else:
    return False
```

```
from bs4 import BeautifulSoup
import requests

gloab_links = set()
def get_links(url):
    response = requests.get(url)
    data = response.text
    soup = BeautifulSoup(data, 'lxml')

    links = set()
    for link in soup.find_all('a'):
        link_url = link.get('href')
        print(link_url)
        if link_url is not None and link_url.startswith('https://www.zehabesha.com/amharic/'):
            links.add(link_url + '\n')

    write_to_file(links)
    return links


def write_to_file(links):
    with open('zehabesha_1.txt', 'a') as f:
        f.writelines(links)


def get_all_links(url):
    for link in get_links(url):
      try:
        if link not in gloab_links\
         and link != 'http://www.litespeedtech.com/error-page'\
         and  link.startswith('https://www.zehabesha.com/amharic/'):
          gloab_links.add(link)
          get_all_links(link)
      except:
        continue


r = 'https://www.zehabesha.com/amharic/'
write_to_file([r])
get_all_links(r)
```
